```python
import numpy as np
import pandas as pd

dat=pd.read_table('cps09mar.txt', header=None)
#rename all the columns
dat.columns=['age', 'gender', 'hisp', 'edu', 'earning', 'hours', 'week', 'union', 'uncov', 'region', 'race', 'marital']
#get experience and its square
dat['exp']=dat['age']-dat['edu']-6
dat['exp^2']=dat['exp']*dat['exp']/100
#get lnwage
dat['wage_per_hour']=dat['earning']/(dat['week']*dat['hours'])
dat['ln_wage']=np.log(dat['wage_per_hour'])
dat.tail()
```

```python
asian=dat.loc[dat['race'] == 4]
single_asian=asian.loc[asian['marital']==7]
single_asian_male=single_asian.loc[single_asian['gender']==0]
single_asian_male_final=single_asian_male.loc[single_asian_male['exp']<=45]
single_asian_male_final['ones']=1
single_asian_male_final.tail()
```

```python
panel_1=single_asian_male_final[[ 'edu', 'exp', 'exp^2', 'ones']].values
beta=np.dot(np.linalg.inv(np.dot(np.transpose(panel_1), panel_1)), np.dot(np.transpose(panel_1), single_asian_male_final[['ln_wage']].values))
beta
```
