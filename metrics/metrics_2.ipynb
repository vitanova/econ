```python
import numpy as np
import pandas as pd

dat=pd.read_table('cps09mar.txt', header=None)
#rename all the columns
dat.columns=['age', 'gender', 'hisp', 'edu', 'earning', 'hours', 'week', 'union', 'uncov', 'region', 'race', 'marital']
#get experience and its square
dat['exp']=dat['age']-dat['edu']-6
dat['exp^2']=dat['exp']*dat['exp']/100
#get lnwage
dat['wage_per_hour']=dat['earning']/(dat['week']*dat['hours'])
dat['ln_wage']=np.log(dat['wage_per_hour'])
dat.tail()
```

```python
aim=dat.loc[(dat['race'] == 1) & (dat['gender']==0) &(dat['hisp']==1)]
aim.tail()
```

```python
def is_it(num, value):
    if num==value:
        return 1
    else:
        return 0
aim['northeast']=aim.apply(lambda row: is_it(1, row['region']), axis=1)
aim['south']=aim.apply(lambda row: is_it(3, row['region']), axis=1)
aim['west']=aim.apply(lambda row: is_it(4, row['region']), axis=1)
aim['married']=aim.apply(lambda row: max(is_it(1, row['marital']), is_it(2, row['marital']), is_it(3, row['marital'])), axis=1)
aim['widowed_divorced']=aim.apply(lambda row: max(is_it(4, row['marital']), is_it(5, row['marital'])), axis=1)
aim['separated']=aim.apply(lambda row: is_it(6, row['marital']), axis=1)
aim['ones']=1
aim
```

```python
panel_1=aim[[ 'edu', 'exp', 'exp^2', 'northeast', 'south', 'west', 'married', 'widowed_divorced', 'separated', 'ones']].values
beta=np.dot(np.linalg.inv(np.dot(np.transpose(panel_1), panel_1)), np.dot(np.transpose(panel_1), aim[['ln_wage']].values))
beta
```
