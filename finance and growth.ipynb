```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

alpha=0.5
lamda=4
B=0.1
theta=0.5
gamma=20

pi=lambda alpha: (1-alpha)*alpha**((1+alpha)/(1-alpha))
phi=lambda alpha: (1-alpha)*alpha**((2*alpha)/(1-alpha))
def rnd(mu):
    return gamma*mu*(pi(alpha)-mu/lamda**2)
def icc(mu):
    return B*gamma*mu**2/(theta*lamda**2)+phi(alpha)
def new(mu):
    return gamma*mu*(pi(alpha)-mu/lamda**2)-B*gamma*mu**2/(theta*lamda**2)
def phiphi(mu):
    return phi(alpha)+mu-mu

Gamma=[5,10,20]
fig, axes=plt.subplots(1,3,figsize=(12,6))
Mu=np.linspace(0,1,100)
for i in range(3):
    gamma=Gamma[i]
    axes[i].plot(Mu, rnd(Mu),lw=2, alpha=0.8, label='R&D')
    axes[i].plot(Mu, icc(Mu),lw=2, alpha=0.8, label='ICC')
    axes[i].legend()
    t=r'$\gamma={}$'.format(gamma)
    axes[i].set_title(t)
plt.show()
```
