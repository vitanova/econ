```python
%matplotlib inline
#I am going to apply three methods of sde numerical simulation according to wikipedia:
#https://en.wikipedia.org/wiki/Stochastic_differential_equation
#https://en.wikipedia.org/wiki/Euler%E2%80%93Maruyama_method
#namely: euler-maruyama; runge-kutta; milstein
#population growth equation
#dY_t=r*Y_t*dt+alpha*Y_t*dW_t, Y_0=IC
import numpy as np 
import matplotlib.pyplot as plt

tBegin = 0
tEnd = 2
dt = .00001

t = np.arange(tBegin, tEnd, dt)
N = t.size 
IC = 1
r=0.01
alpha=0.1

sqrtdt = np.sqrt(dt)
y = np.zeros(N)
z = np.zeros(N)
x = np.zeros(N)
y[0] = IC #euler-maruyama
z[0] = IC #runge-kutta
x[0] = IC #milstein
def rn():
    return np.random.normal(loc=0.0, scale=1.0)
for i in range(1, N):
    dw=sqrtdt*float(rn())
    y[i]=y[i-1]+r*y[i-1]*dt+alpha*y[i-1]*dw
    z[i]=z[i-1]+r*z[i-1]*dt+alpha*z[i-1]*dw
    +0.5*alpha*(r*z[i-1]*dt+alpha*z[i-1]*sqrtdt)*((dw)*2-dt)/sqrtdt
    x[i]=x[i-1]+r*x[i-1]*dt+alpha*x[i-1]*dw
    +0.5*alpha*x[i-1]*alpha*((dw)*2-dt)
ax = plt.subplot(111)
ax.plot(t, y, color='red', lw=1, alpha=0.6, label='euler-maruyama')
ax.legend()
plt.show()
ax = plt.subplot(111)
ax.plot(t, y, color='blue', lw=1, alpha=0.6, label='runge-kutta')
ax.legend()
plt.show()
ax = plt.subplot(111)
ax.plot(t, y, color='green', lw=1, alpha=0.6, label='milstein')
ax.legend()
plt.show()
```
