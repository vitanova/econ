```python
"""
I am going to draw the core picture in the paper IPCs with GSCs, where supply chain goes global.
"""

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
import numpy as np

#step1 draw 2 ellipses
fig=plt.figure(figsize=(11,11))
ax=fig.add_subplot(111, aspect='equal')
ax.set_xlim(-0.1, 1.0)
ax.add_patch(
    patches.Ellipse(
        (0.5, 0.25),
        0.6,
        0.4,
        color='grey',
        alpha=0.15
    )
)
ax.add_patch(
    patches.Ellipse(
        (0.5, 0.65),
        0.6,
        0.4,
        color='grey',
        alpha=0.3
    )
)
#draw north supply chains
ax.plot([0.3, 0.7], [0.75, 0.75], color='black', lw=1.5)
ax.plot([0.3, 0.3], [0.74, 0.76], color='black', lw=1.5)
ax.annotate(r"$stage\ 0$", xy=(0.25,0.71), fontsize=12)
ax.annotate(r"$\rightarrow$", xy=(0.35,0.71), fontsize=15)
ax.plot([0.7, 0.7], [0.74, 0.76], color='black', lw=1.5)
ax.annotate(r"$stage\ 1$", xy=(0.65,0.71), fontsize=12)
ax.annotate(r"$\rightarrow$", xy=(0.55,0.71), fontsize=15)
ax.plot([0.45, 0.45], [0.74, 0.76], color='black', lw=1.5)
ax.annotate(r"$stage\ s^*$", xy=(0.41,0.71), fontsize=12)
ax.annotate(r"$\overbrace{\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }^{north\ unskilled}$", 
            xy=(0.3, 0.76), fontsize=13)
ax.annotate(r"$\overbrace{\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }^{north\ skilled}$", 
            xy=(0.45, 0.76), fontsize=13)

ax.plot([0.3, 0.7], [0.65, 0.65], color='black', lw=1.5)
ax.plot([0.3, 0.3], [0.64, 0.66], color='black', lw=1.5)
ax.annotate(r"$stage\ 0$", xy=(0.25,0.61), fontsize=12)
ax.plot([0.7, 0.7], [0.64, 0.66], color='black', lw=1.5)
ax.annotate(r"$stage\ 1$", xy=(0.65,0.61), fontsize=12)
ax.plot([0.45, 0.45], [0.64, 0.66], color='black', lw=1.5)
ax.annotate(r"$stage\ s^*$", xy=(0.41,0.61), fontsize=12)

ax.plot([0.3, 0.7], [0.55, 0.55], color='black', lw=1.5)
ax.plot([0.3, 0.3], [0.54, 0.56], color='black', lw=1.5)
ax.plot([0.7, 0.7], [0.54, 0.56], color='black', lw=1.5)
ax.plot([0.45, 0.45], [0.54, 0.56], color='black', lw=1.5)

#then south
ax.plot([0.3, 0.7], [0.35, 0.35], color='black', lw=1.5)
ax.plot([0.3, 0.3], [0.34, 0.36], color='black', lw=1.5)
ax.annotate(r"$stage\ 0$", xy=(0.25,0.31), fontsize=12)
ax.plot([0.7, 0.7], [0.34, 0.36], color='black', lw=1.5)
ax.annotate(r"$stage\ 1$", xy=(0.65,0.31), fontsize=12)
ax.annotate(r"$\rightarrow$", xy=(0.45,0.31), fontsize=15)
ax.annotate(r"$\overbrace{\ \ \ \ \ \ \ \ \ \ \\  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }^{south\ unskilled}$", 
            xy=(0.3, 0.36), fontsize=13)

ax.plot([0.3, 0.7], [0.25, 0.25], color='black', lw=1.5)
ax.plot([0.3, 0.3], [0.24, 0.26], color='black', lw=1.5)
ax.annotate(r"$stage\ 0$", xy=(0.25,0.21), fontsize=12)
ax.plot([0.7, 0.7], [0.24, 0.26], color='black', lw=1.5)
ax.annotate(r"$stage\ 1$", xy=(0.65,0.21), fontsize=12)

ax.plot([0.3, 0.7], [0.15, 0.15], color='black', lw=1.5)
ax.plot([0.3, 0.3], [0.14, 0.16], color='black', lw=1.5)
ax.plot([0.7, 0.7], [0.14, 0.16], color='black', lw=1.5)

#now, add arrows between two regimes
ax.add_patch(patches.FancyArrowPatch(
    (0.2, 0.25),
    (0.2, 0.65),
    connectionstyle='arc3, rad=-0.5',
    mutation_scale=20,
    fill=False))
ax.add_patch(patches.FancyArrowPatch(
    (0.8, 0.65),
    (0.8, 0.25),
    connectionstyle='arc3, rad=-0.5',
    mutation_scale=20,
    fill=False))
ax.annotate("innovation by north", xy=(0.1, 0.45), fontsize=17)
ax.annotate("imitation by south", xy=(0.7, 0.45), fontsize=17)
ax.annotate(r"$\xi_N\ north$", xy=(0.6, 0.5), fontsize=17)
ax.annotate(r"$\xi_S\ south$", xy=(0.6, 0.1), fontsize=17)

#eliminate sticks and spines
ax.set_xticks([])
ax.set_yticks([])
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.set_title("IPCs with LSCs")
```
