```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

rho=0.9
gamma=2

def newp(p):
    return (rho*(1-p)*(1-p)*(gamma*(1+p)**2-(1-p)**2)+1)/2

P=np.linspace(0.01, 0.99, 100)
PP=[]
for p in P:
    pp=newp(p)
    PP.append(pp)

fig, ax=plt.subplots(figsize=(5, 5))
ax.plot(P, P, color='blue')
ax.plot(P, PP, color='red')
ax.set_ylim(0, 1)
plt.show()
```
