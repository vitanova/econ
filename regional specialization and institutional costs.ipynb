```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

alpha1=0.8
alpha2=0.2

T=np.linspace(0, 3.5, 300)
C1=[]
C2=[]
for t in T:
    temp1=alpha1+(1-alpha1)*t
    temp2=alpha2+(1-alpha2)*t
    C1.append(temp1)
    C2.append(temp2)

fig, ax = plt.subplots(figsize=(6, 5))
ax.plot(T, C1, color="blue", lw=1.5, linestyle="-", label=r"$c_1=\alpha_1+(1-\alpha_1)t$")
ax.plot(T, C2, color="green", lw=2.5, linestyle="-.", label=r"$c_2=\alpha_2+(1-\alpha_2)t$")
ax.plot([1, 1], [0, 1], color='black', lw=1, linestyle="--")
ax.plot([0, 1], [1, 1], color='black', lw=1, linestyle="--")
ax.set_ylim(0,2)
ax.set_xlim(0, 2)
ax.scatter([1,], [1,], 20, color='black')
ax.annotate(r"$t_1=1, c_1=c_2=1$", xy=(1, 1), xycoords='data', xytext=(+10, +30), 
           textcoords='offset points', fontsize=16,
           arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.3"))
ax.set_xlabel('institutional parameter $t$')
ax.set_ylabel('marginal cost $c$')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.legend()
plt.show()
```

```python
t_1=1
t_2=2.5
p_A=1
def w(t):
    if t==t_1:
        return np.sqrt((alpha1+(1-alpha1)*t_2)/(alpha1+(1-alpha1)*t_1))*p_A
    elif t>=t_2:
        return p_A
    elif t>t_1:
        return np.sqrt((alpha1+(1-alpha1)*t_2)/(alpha1+(1-alpha1)*t))*p_A
    else:
        return np.sqrt((alpha2+(1-alpha2)*t_1)/(alpha2+(1-alpha2)*t))*w(t_1)
W=[]
for t in T:
    W.append(w(t))
def w_1(t):
    return np.sqrt((alpha1+(1-alpha1)*t_2)/(alpha1+(1-alpha1)*t))*p_A
T_1=np.linspace(0, 1, 50)
W_1=[]
for t in T_1:
    W_1.append(w_1(t))

fig, ax=plt.subplots(figsize=(8,5))
ax.plot(T, W, color='blue', lw=2, linestyle='-', label=r"$w=w(t)$")
ax.plot(T_1, W_1, color='black', lw=1.5, linestyle='--')
ax.plot([0, 2.5], [1, 1], color='grey', lw=1.5, linestyle='--')
ax.plot([1, 1], [0, w(1)], color='black', lw=1.5, linestyle='--')
ax.plot([2.5, 2.5], [0, 1], color='grey', lw=1.5, linestyle='--')
ax.scatter([1,], [w(1),], 20, color='black')
ax.scatter([2.5,], [1,], 20, color='black')
ax.set_xlim(0, 3.5)
ax.set_ylim(0.5, 2.5)
ax.set_xlabel('institutional parameter $t$')
ax.set_ylabel('regional wage $w$')
ax.annotate(r"$(t_1, w_2=w_1)$", xy=(1, w(1)), xycoords='data', xytext=(+10, +30), 
           textcoords='offset points', fontsize=16,
           arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.3"))
ax.annotate(r"$(t_2, w_1=p^A)$", xy=(2.5, 1), xycoords='data', xytext=(+10, +30), 
           textcoords='offset points', fontsize=16,
           arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.3"))
ax.annotate("Type-M2 Regions", xy=(0.13, 0.75), fontsize=11)
ax.annotate("Type-M1 Regions", xy=(1.4, 0.75), fontsize=11)
ax.annotate("Type-A Regions", xy=(2.7, 0.75), fontsize=11)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.legend()
plt.show()
```
